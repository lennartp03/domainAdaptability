```python
import kornia as K
import kornia.feature as KF
import torch
import cv2
import matplotlib.pyplot as plt
```

```python
%%capture
fname1 = "./hpatches-sequences-release/i_ajuntament/1.ppm"  # Change the file extension to .ppm
fname2 = "./hpatches-sequences-release/i_ajuntament/5.ppm"  # Change the file extension to .ppm

img1 = cv2.imread(fname1, cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread(fname2, cv2.IMREAD_GRAYSCALE)

# Resize the images if needed
img1 = cv2.resize(img1, (600, 375))
img2 = cv2.resize(img2, (600, 375))

# Initialize the LoFTR matcher
matcher = KF.LoFTR(pretrained="outdoor")

input_dict = {
    "image0": img1,  # No need to convert to grayscale, as LoFTR works on color images
    "image1": img2,  # No need to convert to grayscale, as LoFTR works on color images
}

with torch.inference_mode():
    correspondences = matcher(input_dict)
```
